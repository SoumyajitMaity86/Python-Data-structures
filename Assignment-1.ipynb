```python
#32.Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are
#common keys the values should be added together.

def merge(dict1, dict2):
   
    merged = dict1.copy()  # Start with dict1's keys and values by coping in merged
    
    for key, value in dict2.items():
        if key in merged:
            merged[key] +=value  # Add the values if key exists in both dictionaries
        else:
            merged[key] = value   # Add the new key-value pair if key does not exist in merged_dict
    
    return merged
dict1={}
dict2={}
    # Take the first dictionary as input from the user
n= int(input("Enter the number of entries you want to add: "))
for i in range(n):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict1[key] = value
    
    # Take the second dictionary as input from the user
n=int(input("Enter the number of entries you want to add: "))
for i in range(n):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict2[key] = value
    
    # Merge the dictionaries
merged_dict = merge(dict1, dict2)
    
    # Print the merged dictionary
print("Merged dictionary:", merged_dict)
```

```python
#35.Write a code that inverts a dictionary swapping keys and values. Ensure that tha inverted dictionary correctly handles cases
#where multiple keys have the same value by storing the keys as a list in the inverted dictionary.

def invert(original_dict):
    inverted= {}
    for key, value in original_dict.items():
        if value in inverted:
            inverted[value].append(key)
        else:
            inverted[value] = [key]
    return inverted

dict1={}
   # Take the dictionary as input from the user
n= int(input("Enter the number of entries you want to add: "))
for i in range(n):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict1[key] = value

# Invert the dictionary
inverted = invert(dict1)

# Print the inverted dictionary
print("Inverted dictionary:", inverted)
```
